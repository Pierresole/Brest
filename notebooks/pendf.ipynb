```python
import sys
import os
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import endf

# Determine the directory where the notebook is being executed
notebook_dir = Path().resolve()
project_path = notebook_dir.parent
data_path = project_path / 'data'

# Add the project and reconstruction directories to the Python path
if str(project_path) not in sys.path:
    sys.path.append(str(project_path))
if str(data_path) not in sys.path:
    sys.path.append(str(data_path))
    
from parser.parse_mf3 import read_MF3
```

```python
file_path = data_path / 'dop-80160-0.0.pendf' 
energies, xs = read_MF3(file_path, 102)
```

```python
# plot xs array versus energies array in log-log scale
plt.plot(energies, xs)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Energy [eV]')
plt.ylabel('Cross Section [barns]')
plt.title('Cross Section vs Energy')
plt.xlim(1e5, 1e7)
plt.show()
```

```python
mat = endf.Material(file_path)
xs = mat.section_data[3, 102]['sigma']

# plot xs array versus energies array in log-log scale
plt.plot(xs.x, xs.y)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Energy [eV]')
plt.ylabel('Cross Section [barns]')
plt.title('Cross Section vs Energy')
plt.xlim(1e5, 1e7)
plt.show()
```
